### Nightingale Chart (Wind Rose Chart)

Using plotly:  https://plotly.com/python/wind-rose-charts/ 

```python
import plotly.express as px
import pandas as pd

colors = ["#ff412c", "#ff8d80", "#ccd1db", "#c57fff", "#9f29ff"]
values = [52.81, 21.03, 13.13, 7.63, 5.40]
labels = ["Very serious", "Somewhat serious", "Not sure", "Not very serious", "Not at all serious"]

df = pd.DataFrame({"color" : colors, "Percent":values, "Response": labels})


title= "How serious of a problem do you think it is that senior members of the Trump administration discussed highly sensitive military plans\n"
title += "using an unclassified chat application in a group that inadvertently included a journalist? (%)"

fig = px.bar_polar(df, r="Percent", theta="Response", color="Response", color_discrete_sequence = colors)

fig.update_layout(title = title, title_font_size = 7.5)
fig.show()
```

Using pyplot:  https://cookieblues.medium.com/how-to-make-a-nightingale-rose-diagram-only-using-matplotlib-c581441e8cc4 

```python
import numpy as np

colors = ["#ff412c", "#ff8d80", "#ccd1db", "#c57fff", "#9f29ff"]
values = [52.81, 21.03, 13.13, 7.63, 5.40]
labels = ["Very serious", "Somewhat\nserious", "Not sure", "Not very\nserious", "Not at all\nserious"]

def format_pct(x):
    return f"{round(x, 1)}%"

labels = [labels[i] + f"\n{format_pct(values[i])}" for i in range(len(labels))]

n_points = len(values)
inner_radius = 1

x_max = 2*np.pi
x_coords = np.linspace(np.pi / 5 + np.pi/2, x_max + np.pi / 5 + np.pi/2, n_points, endpoint=False)
width = x_max / n_points

plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)
ax.bar(
    x_coords,
    values[::-1],
    width=width,
    bottom=inner_radius,
    color=colors[::-1],
    edgecolor="white",
    linewidth=1
)


text_placement = [13, 13, 10, 15, 12]

for i in range(5):
    ax.text(x_coords[i] - 2 *np.pi / 5, text_placement[-1 * i], labels[-1 *i], ha="center", va="center")


plt.axis("off")


title= "How serious of a problem do you think it is that senior members of the Trump administration, including\n"
title += "Vice President JD Vance and Defense Secretary Pete Hegseth, discussed highly sensitive military plans\n"
title += "using an unclassified chat application in a group that inadvertently included a journalist? (%)"

ax.text(np.pi / 2, 55, title, fontsize = 8, fontweight="bold", ha="center")


plt.show()
```

### Showing subcategories

A total of 5976 U.S. adults were surveyed. Let's first find out how many are in each political party. Time for some linear algebra! 

```python
n = 5976

totals = [52.81, 21.03, 13.13, 7.63, 5.40]

categories = [[75.49, 14.11, 7.06, 2.85, 0.48],  # democrats
          [52.76, 18.87, 15.48, 7.59, 5.30], # independents
          [28.01, 31.74, 16.35, 12.94, 10.95]] # republicans

# val[0] * x + val[1] * y + val[2] * z = total percent 
# find percent of total for x, y, z

matrix = []
for i in range(5):
    cur_row = [categories[j][i] for j in range(3)]
    matrix.append(cur_row)
    
x = np.linalg.solve(matrix[:3], totals[:3])
print(x)

# verify solution
for i in range(5):
    check = sum(x[j] * matrix[i][j] for j in range(3))
    print(check, totals[i])
```

```python
# Version 2 -- stacked bar chart

categories = [[28.01, 31.74, 16.35, 12.94, 10.95], # republicans
              [52.76, 18.87, 15.48, 7.59, 5.30], # independents
              [75.49, 14.11, 7.06, 2.85, 0.48]]  # democrats

proportions = [0.28168625, 0.40996658,  0.30851942]
cmaps = [plt.cm.Reds, plt.cm.Greys, plt.cm.Blues]

fig, ax = plt.subplots()

cur_x = 0
x_ticks = []
for i in range(3):
    cur_bottom = 0
    for j in range(5):
        ax.bar(cur_x, categories[i][j], bottom = cur_bottom, color=cmaps[i](0.9 - (j / 5)),
               width = proportions[i] * 3, align = "edge")
        cur_bottom += categories[i][j]
    x_ticks.append(cur_x + proportions[i] * 3 / 2)
    cur_x += proportions[i] * 3
    
# set labels 
labels = ["Very serious", "Somewhat\nserious", "Not sure", "Not very\nserious", "Not at all\nserious"]

for i in range(3):
    j = 0
    cur_y = 0
    while categories[i][j] > 5: # don't label boxes that are too small
        
        cur_label = labels[j]
        if categories[i][j] < 10:
            cur_label = cur_label.replace("\n", " ")
        
        ax.text(x_ticks[i], cur_y + categories[i][j] / 2, cur_label, ha="center", va="center", color = "white")
        cur_y += categories[i][j]
        j += 1
        if j == 5:
            break
        
ax.set_xticks(x_ticks)
ax.set_xticklabels(["Republicans", "Independents", "Democrats"])
ax.set_yticklabels([])
ax.tick_params(axis="both", length =0)

ax.spines[["left", "right", "top", "bottom"]].set_visible(False)


title= "How serious of a problem do you think it is that senior members of the Trump administration, including\n"
title += "Vice President JD Vance and Defense Secretary Pete Hegseth, discussed highly sensitive military plans\n"
title += "using an unclassified chat application in a group that inadvertently included a journalist? (%)"

ax.set_title(title, fontsize = 8, fontweight="bold")

plt.show()
```

### Data Export

Show new values based on party affiliation first, then ranking

```python
categories = [[28.01, 31.74, 16.35, 12.94, 10.95], # republicans
              [52.76, 18.87, 15.48, 7.59, 5.30], # independents
              [75.49, 14.11, 7.06, 2.85, 0.48]]  # democrats

proportions = [0.28168625, 0.40996658,  0.30851942]

labels = ["Very serious", "Somewhat serious", "Not sure", "Not very serious", "Not at all serious"]
data_dict = {"Category": labels}

parties = ["Republicans", "Independents", "Democrats"]

for i in range(3):
    data_dict[parties[i]] = [categories[i][j] * proportions[i] for j in range(5)]
    
df_export = pd.DataFrame(data_dict)
df_export.to_csv("survey-data-tableau2.csv", index = False)
    

```
