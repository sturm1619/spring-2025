# Part 1: Visualizing Amounts

## Plot (a): Fictitious Company

I decided to generate my own dataset for the visualizations. This dataset was generated using the model of a fictitious company. In this model, our company has $N=6$ departments, and the total number of employees $f(x)$ a function of the number of stores $x$ and the number of products $y$, such that

$$ f(x) = \sum_{\textbf{a}} (a_nx + \epsilon_n) $$

where $\textbf{a}$ is the ordered set where the $n$th entry is the weight corresponding to a specific department in the company, and $\epsilon_n$ is an error value unique to each department. Therefore, $a_nx + \epsilon_n$ is the number of employees in the $n$th department. The following table summarizes the weigths ($\textbf{a})$:

|Index $n$|Department      | Weight ($a_n$) |
|---------|----------------|----------------|
|1        |Product Desg.   |$2$             |
|2        |Marketing       |$2$             |
|3        |Retail          |$6$             |
|4        |Accounting      |$2$             |
|5        |Logistics       |$4$             |
|6        |Human Rersources|$2$             |

```python
number_of_stores: int = 5

from math import ceil
from random import gauss

company_data = pd.DataFrame(
    [number_of_stores * department_weight + ceil(gauss(0,3)) for department_weight in (2,2,6,2,4,2)],
    index=["Product Design", "Marketing", "Retail", "Accounting", "Logistics", "HR"],
    columns=["Employees"], dtype="uint16"
)

company_data
```

```python
from math import sqrt, floor

total_employees = company_data.Employees.sum()
default_row, default_column = floor(sqrt(total_employees)), ceil(sqrt(total_employees))
squares_in_waffle = default_row * default_column

department_data = {
    number : {
        'values': [(number_of_employees := company_data.loc[department, "Employees"]), squares_in_waffle - number_of_employees],
        'colors': ['tab:orange', 'tab:gray'],
        'title': {'label': f"{number_of_employees} employees", 'fontweight': 'bold', "color": "tab:orange"},
        'vertical': True, 'starting_location': 'SW'
    } for number, department in enumerate(company_data.index, start=231)
}

fig = plt.figure(
    FigureClass = Waffle,
    plots = department_data,
    rows = default_row, columns = default_column,
    figsize=(10,6)
)
for title, axes in zip(company_data.index, fig.get_axes()):
    axes.text(0.55, 1.1, title, ha="center")
    
plt.suptitle("Number of employees per department", fontweight="bold")
plt.show()
```

## Part (b): Demographics

The analytic team of our fictitious company decided to review the sells of the last quarter and see the how their products are selling. All of the product follow the same concept, which is targetting a very specific demographic. Since there are many demographic groups, they plotted everything into one heatmap.

The data was generated as follows: Let $f: \{F, M\} \times \{W, B, A, N, H\} \times Z \rightarrow R$. The number of total sales $f(x,y,z)$ in a demographic group identified by gender $x \in \{F, M\}$, ethnicity $y \in \{W,B,A,N,H\}$, and age $z \in Z$  is given by

$$ f(x,y,z) = 10(100 \cdot g(x)\cdot h(y) - (z - 8)^{1.2} + \epsilon) $$

Where 
- $g: \{F, M\} \rightarrow R$ such that $q(F) = \frac{4}{5}$, $q(M) = \frac{1}{3}$,
- $h: \{W,B,A,N,H\} \rightarrow R$ such that $h(W) = h(B) = 1, h(H) = \frac{4}{5}, h(A) = \frac{3}{5}, h(N) = \frac{3}{10}$,
- $\epsilon$ is an error value unique to each demographic.

```python
from math import ceil
from random import gauss
from itertools import product

def calculate_sales(gender, ethnicity, age) -> int:
    weight_per_gender = {"F":1, "M": 1/3}
    weight_per_ethnicity = {"W": 1, "B": 1, "A": 0.6, "H": 0.8, "N" : 0.3}
    
    sales: float = \
        (1000 - (age - 8)**(1.8) + gauss(0,2)) \
        * weight_per_gender[gender] * weight_per_ethnicity[ethnicity]
    return ceil(sales) if sales > 0 else 0

demography = list(product("FM", "WBHAN"))

def generate_data() -> dict[str, list[float]]:
    data = {"ages": list(range(8,56))}
    for gender, ethnicity in demography:
        data[f"{gender}{ethnicity}"] = \
            [calculate_sales(gender, ethnicity, age)/1000 for age in data["ages"]]
    return data

data = generate_data()

sales_data = pd.DataFrame(
    data, index=data["ages"],
    columns=[f"{gender}{ethnicity}" for gender, ethnicity in demography]
)
sales_data.T # Transposed for easier visualization
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(
    sales_data,
    axes=ax, cmap="spring",
    annot=True, cbar=False
)
#ax.set_xticks([]); ax.set_yticks([])
ax.set_title("Sales per demography (in thousands)");
```

# Part 2: Time charts

## Polar chart

The company wants to see the evolution of the sales throughout the year, so they decided to plot it into a polar time chart.

The data was generated as follows:

Let $f: Z \rightarrow R$. The number of sales $p(x)$ in a given month $x$ is given by

$$ p(x) = \left|\sin \left(\frac{x}{4}\right)\right| \sum_A f(a) $$.

Where $\sum_A f(a)$ is the sum of the sales per demographic group as described in the last plot.

```python
from math import ceil, sin, pi
years = 2; months = list(range(1, years * 12 + 1))

retail_data = pd.DataFrame(
    [ceil(sum(sum(list) for list in generate_data().values()) * abs(sin(month/4))) for month in months],
    index=months,
    columns=["sells"])

# Data transformation to help plotting
from datetime import date
def to_month(x: int):
    mapped = x % 12
    return mapped if mapped > 0 else 12

retail_data["month"] = [date(2000, month, 1).strftime("%B") for month in retail_data.index.map(to_month)]
retail_data["polar"] = (retail_data.index.map(to_month) -1) * pi/6

retail_data.T
```

```python
fig, ax = plt.subplots(subplot_kw={'projection':'polar'})
ax.plot(retail_data.polar, retail_data.sells);
ax.grid(visible = False, axis = 'x')

xticks = (np.arange(1,13) - 1) * pi/6
ax.set_xticks((np.arange(1,13) - 1) * pi/6);
ax.set_xticklabels([date(2000, month, 1).strftime("%B") for month in range(1,13)]);

yticks = ax.get_yticks()
ax.set_yticks(yticks[1:len(yticks):2]);
ax.set_title("Sells per month in the last 2 years");
#ax.annotate("Best month", (pi, retail_data.sells.max() - 100))
```

## Part (b): Sales per Demography

The C-level executives want to see the demographic trends across the months.

For a demographic group $(x,y,z)$,

$$ g(x,y,z) = \left|\sin \left(\frac{x}{4}\right)\right| f(x,y,z)$$

from math import ceil, sin, pi
from itertools import productis the total sales in that month. Then $g(x,y) = \sum_A f(x,y,a)$ is the cumulative sum of the total sells in a demograpic group across all its ages, where $a$ corresponds to a specific age group.

```python
from math import ceil, sin, pi
from random import gauss
from itertools import product
from datetime import datetime

def calculate_sales(gender, ethnicity, age) -> int:
    weight_per_gender = {"F":1, "M": 1/3}
    weight_per_ethnicity = {"W": 1, "B": 1, "A": 0.6, "H": 0.8, "N" : 0.3}
    
    sales: float = \
        (1000 - (age - 8)**(1.8) + gauss(0,2)) \
        * weight_per_gender[gender] * weight_per_ethnicity[ethnicity]
    return ceil(sales) if sales > 0 else 0

def to_month(x: int):
    mapped = x % 12
    return mapped if mapped > 0 else 12

demography = list(product("FM", "WBHAN"))
data = pd.DataFrame()
for month in range(1, 13):
    for gender, ethnicity in demography:
        data.loc[month, f"{gender}{ethnicity}"] = \
        ceil(sum(calculate_sales(gender, ethnicity, year) for year in range(8,56)) * abs(sin(month/4)))

data
```

```python
colors = plt.cm.Reds(data.sum() / np.max(data.sum()))
#print(colors)
fig, ax = plt.subplots(figsize=(10,6))

plt.stackplot(
    data.index, data.T, labels=data.columns,
    colors=colors, edgecolor="black", linewidth=0.3
)
ax.set_title("Sells per demography accross months")
ax.set_xlabel("Months")
ax.set_ylabel("Sells")
ax.set_xticks(range(1,13))
ax.set_xticklabels(list(datetime(1, month, 1).strftime("%b") for month in range(1,13)))
ax.legend();
#ax.spines["lef
```
