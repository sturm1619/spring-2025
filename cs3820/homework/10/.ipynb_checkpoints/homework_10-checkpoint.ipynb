```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Distributions

## Academic Pressure (Dot plot)

Honestly, didn't like how this one looked. Academic Pressure seems to be an _ordinal_ variable, so treating it like amounts would yield a better graph.

```python
student_depression = pd.read_csv("student_depression_dataset.csv")
academic_pressure_summary = student_depression \
    ["Academic Pressure"] \
    .value_counts()

student_depression.head();
```

```python
scale = int(5 * 100)
x_vals = []
y_vals = []
for i in range(int(max(academic_pressure_summary.index)) + 1):
    y_count = scale
    for j in range(academic_pressure_summary.loc[i] // scale):
        x_vals.append(i)
        y_vals.append(y_count)
        y_count += scale
 
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(x_vals, y_vals)

ax.hlines(y=0, xmin=.5, xmax=5.5, color="lightgray", linewidth=1)
for i in range(0, 6):
    ax.vlines(x = i + .5, ymin=0, ymax= 7000, color="lightgray", linewidth=1)

ax.spines[:].set_visible(False)
ax.tick_params(which="both", length=0)

ax.set_xlim(0, 6)
ax.set_xticks(range(1,6))

ax.set_xlabel("Stress level")
ax.set_ylabel("Number of students")
fig.suptitle("Most students find school very stressing");
```

## Caffeine Intake (Dot plot 2)

```python
sleep_data = pd.read_csv("sleeptime_prediction_dataset.csv")

fig, axs = plt.subplots(2, 1)
sns.stripplot(x = sleep_data.WorkHours, alpha=0.6, ax = axs[1], color="tab:blue")
sns.kdeplot(sleep_data.WorkHours, fill=True, ax = axs[0])

for ax in axs:
    ax.spines[["top", "bottom", "left", "right"]].set_visible(False)
    ax.set_xlim(sleep_data.WorkHours.min(), sleep_data.WorkHours.max())
    ax.set_yticks([])
    ax.set_ylabel("")
    ax.tick_params(axis="both", length = 0)
axs[0].set_xticks([])
axs[0].set_xlabel("")
axs[1].set_ylim(-0.15, 1)
axs[1].zorder = -10

fig.suptitle("The deviation of hours at work per day is very uniform")

plt.tight_layout(h_pad=-16)
plt.show()
```

# Relationships

## Sleep and covariates (Heatmap)

```python
sleep_time_corr = sleep_data.corr()
mask = np.triu(np.ones_like(sleep_time_corr, dtype="bool"), k=1);

# Remove duplicated values
sleep_time_corr.drop(index="WorkoutTime", columns="SleepTime", inplace=True)
mask = mask[1:mask.shape[0], 1:mask.shape[1]]

#Plot
ax = sns.heatmap(
    sleep_time_corr.T, mask=mask.T,
    vmin=-1, vmax=1, cmap='coolwarm',
    annot=True
)

ax.tick_params(axis='both', length=0)
ax.set_title(
    "Workout and Caffeine have small effects on sleep! D:",
    weight="bold"
);
```

## Sleep and Covariates (Parallel Coordinates)

Not my favorite...

```python
from matplotlib.colors import TABLEAU_COLORS

caffeine_adjusted = sleep_data[sleep_data.CaffeineIntake < 50]
caffeine_adjusted = sleep_data.loc[:9]
covariates = caffeine_adjusted.columns

# Normalizing Dataset
y_val_lists = []
for covariate in covariates:
    y_vals1 = list(caffeine_adjusted[covariate])

    # normalize values
    max1 = max(y_vals1)
    min1 = min(y_vals1)
    y_vals1_norm = [(y - min1) / (max1 - min1) for y in y_vals1]
    y_val_lists.append(y_vals1_norm)

# Creating Line Chart
fig, ax = plt.subplots()
for i in range(len(y_vals1)):
    ax.plot(
        list(range(len(covariates))),
        [y_val_lists[j][i] for j in range(len(covariates))]#,
        #c=list(TABLEAU_COLORS.keys())[i]
    )
    
ax.set_xticks(list(range(len(covariates))))
ax.set_xticklabels(["Workout", "Reading", "Phone", "Work", "Caffeine", "Relax", "Sleep"])
ax.set_yticklabels([])

ax.tick_params(axis="both", length = 0)
ax.spines[["left", "right", "top", "bottom"]].set_visible(False)

ax.grid(axis="x", linestyle = "--")
fig.suptitle("Heaven help us: Statisticians don't know how to read this")
ax.set_title("We couldn't find fitting descriptive title...", fontsize=10)
plt.show()
```
