# Relationships 1

This tutorial will cover some basics about scatter plots and bubble plots

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Scatter Plots

Let's generate some random values between 1 and 10:

```python
x_vals = list(range(1, 11))
y_vals = np.random.rand(1, 10) * 10
```

We can generate a scatter plot by simply calling `scatter`:

```python
fig, ax = plt.subplots()
ax.scatter(x_vals, y_vals[0])
```

We can also generate a second data set to add to the graph:

```python
y_vals2 = np.random.rand(1, 10) * 10

fig, ax = plt.subplots()
ax.scatter(x_vals, y_vals[0])
ax.scatter(x_vals, y_vals2[0])
```

Use the `marker` attribute to change from circles to a different shape (can help with distinguishing data sets). Common markers include:
* "s" for square
* "<", ">", "^", and "v" are all triangles (pointing different directions)
* "+" for plus signs 
* "x" for x's 

Additional markers can be found here:  https://matplotlib.org/stable/api/markers_api.html

```python
fig, ax = plt.subplots()
ax.scatter(x_vals, y_vals[0])
ax.scatter(x_vals, y_vals2[0], marker = "s")
```

Some markers may overlap, so changing the `alpha` value for the colors can be helpful:
* `alpha` = 1 is fully opaque, and `alpha` = 0 is fully transparent

```python
fig, ax = plt.subplots()
ax.scatter(x_vals, y_vals[0], alpha = 0.7)
ax.scatter(x_vals, y_vals2[0], alpha = 0.7, marker = "s")
```

The `color` attribute can be used to adjust the marker color. Other helpful marker settings include 
* Setting the `edgecolors` to adjust the marker edges (`face` is the same as the face color, and `none` removes the boundary)
* Setting the `linewidths` of marker boundaries (default is 1.5) 

See https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.scatter.html for more details

```python
fig, ax = plt.subplots()
ax.scatter(x_vals, y_vals[0], alpha = 0.7, edgecolors="black")
ax.scatter(x_vals, y_vals2[0], alpha = 0.7, marker = "s", edgecolors="black")
```

If desired, a line of "best fit" can be included to illustrate trends in the scatter plot
* `np.polyfit` can be used to compute the least-squares regression (as well as higher-order best fit polynomials)

Let's generate some new data that's roughly linear:

```python
y_vals = [i + 0.5 - np.random.rand() for i in range(1, 11)]

fig, ax = plt.subplots()
ax.scatter(x_vals, y_vals)
```

```python
# best fit line
best_fit = np.polyfit(x_vals, y_vals, deg=1) # returns an array containing [slope, intercept]
line_coords = [best_fit[0] * i + best_fit[1] for i in x_vals]
print(best_fit)

fig, ax = plt.subplots()
ax.scatter(x_vals, y_vals)
ax.plot(x_vals, line_coords, color="black", alpha = 0.8)
```

But what if the relationship is non-linear? `np.polyfit` can also be used for regression polynomials of higher degree. You can also "guess" other functions and apply regression methods to fit them to the data. Ideally you should have some statistical reasons behind the function you choose. 

There are also methods that do not involve "guessing" the right function beforehand. One popular method available in Python is the `CubicSpline` from `scipy`. It fits a piecewise cubic polynomial to the data that is everywhere twice-continously differentiable (for you calculus nerds out there). 
* https://docs.scipy.org/doc/scipy-1.15.1/reference/generated/scipy.interpolate.CubicSpline.html



```python
from scipy.interpolate import CubicSpline

x = np.arange(10)
y = np.sin(x)
cs = CubicSpline(x, y)
xs = np.arange(-0.5, 9.6, 0.1)
fig, ax = plt.subplots(figsize=(6.5, 4))
ax.plot(x, y, 'o', label='data')
ax.plot(xs, cs(xs), label="S")
ax.set_xlim(-0.5, 9.5)
ax.legend(loc='lower left', ncol=2)
plt.show()
```

```python
# More random data

data_vals = []
for i in range(100):
    x = np.random.rand() * 10
    y = (x**3 - 13*x**2 + 47*x - 35) / 5 + (2 * np.random.rand() - 1)
    data_vals.append((x, y))
    
data_vals.sort()
x_vals = [x[0] for x in data_vals]
y_vals = [y[1] for y in data_vals]    

    
cs = CubicSpline(x_vals, y_vals)
xs = np.arange(x_vals[0], x_vals[-1], 0.1)
fig, ax = plt.subplots()
ax.plot(x_vals, y_vals, 'o', label='data')
ax.plot(xs, cs(xs), label="S")
# ax.set_xlim(-0.5, 9.5)
ax.legend(loc='upper left', ncol=2)
plt.show()

```

### Bubble Plots

A scatter plot can easily be converted in to a <b>bubble plot</b> by adding an additional weight to scale marker sizes
* Parameter `s` in the `scatter` function controls the marker size (can be a single value, or an array of values)
* Marker size should be given in `points ** 2` (typographic points are 1/72 in.)


```python
x_vals = list(range(1, 11))
y_vals1 = [1] * 10

s1 = list(range(100, 1001, 100))
print(s1)

fig, ax = plt.subplots()
ax.scatter(x_vals, y_vals1, s1)
```

As `s` values increase linearly, the bubble size will scale according to area and not to radius / diameter (see <i>Better Data Visualizations</i>, page 123). 
* Each red dot represents a doubling in size:

```python
x_vals = list(range(1, 11))
y_vals1 = [1] * 10

s1 = list(range(100, 1001, 100))
print(s1)

colors = ["tab:blue"] * 10
colors[0] = "red"
colors[1] = "red"
colors[3] = "red"
colors[7] = "red"

fig, ax = plt.subplots()
ax.scatter(x_vals, y_vals1, s1, color = colors)
```

We can also generate a legend describing the dot sizes:

```python
x_vals = list(range(1, 11))
y_vals1 = [1] * 10

s1 = list(range(100, 1001, 100))
print(s1)

colors = ["tab:blue"] * 10
colors[0] = "red"
colors[1] = "red"
colors[3] = "red"
colors[7] = "red"

fig, ax = plt.subplots()
plot = ax.scatter(x_vals, y_vals1, s1, color = colors)

# produce a legend with a cross-section of sizes from the scatter
handles, labels = plot.legend_elements(prop="sizes", color="tab:blue")
legend2 = ax.legend(handles[::2], labels[::2], loc="upper center", ncols = 5, borderpad=1.6, bbox_to_anchor=(0.5,0.95))

# Remove boundary and y values
ax.spines[['left', 'right', 'top', 'bottom']].set_visible(False)
ax.set_yticks([])
ax.set_xticks([])
```

### Example

Data set:  https://www.kaggle.com/datasets/valakhorasani/gym-members-exercise-dataset



```python
df = pd.read_csv("gym_members_exercise_tracking.csv")

df.head()
```

Let's select all of the Cardio workouts, and make an initial scatter plot by session duration against calories burned:
* Presumably highly correlated events!

```python
df_cardio = df[df["Workout_Type"] == "Cardio"]

df_cardio.plot(kind="scatter", x = "Session_Duration (hours)", y = "Calories_Burned")
```

For fun, let's scale each workout by water intake:

```python
print(min(df_cardio["Water_Intake (liters)"]), max(df_cardio["Water_Intake (liters)"]))
```

```python
df_cardio.plot(kind="scatter", x = "Session_Duration (hours)", y = "Calories_Burned", s = df_cardio["Water_Intake (liters)"] * 20, alpha = 0.7)
```

If we want to see greater distinction among the dot sizes, we can first "normalize" the weights to be values between 0 and 1 and scale from there:
* Apply the formula x' = (x - min) / (max - min)

```python
min_water, max_water = min(df_cardio["Water_Intake (liters)"]), max(df_cardio["Water_Intake (liters)"])
df_water = (df_cardio["Water_Intake (liters)"] - min_water) / (max_water - min_water)

df_cardio.plot(kind="scatter", x = "Session_Duration (hours)", y = "Calories_Burned", s = 10 + df_water * 100, alpha = 0.7)
```

If desired, we can add a legend to describe what the dot sizes mean:

```python
fig, ax = plt.subplots()


plot = ax.scatter(x = df_cardio["Session_Duration (hours)"], y = df_cardio["Calories_Burned"], s = 10 + df_water * 100, alpha = 0.7)
ax.set_xlabel("Session Duration (hours)")
ax.set_ylabel("Calories Burned")

# produce a legend with a cross-section of sizes from the scatter
handles, labels = plot.legend_elements(prop="sizes", alpha=0.7, color="tab:blue")

# Rescale labels to actual values
labels = [round((int(l[l.find('{')  + 1: l.find('}')]) - 10) / 100 * (max_water - min_water) + min_water, 2) for l in labels ]

legend2 = ax.legend(handles[::3], labels[::3], loc="upper left", title="Water (L)")
```

Those particular values in the legend seem unhelpful (though 1.5 is the min value). We can always plot our own dots sizes directly with custom labels. 

```python
fig, ax = plt.subplots()

plot = ax.scatter(x = df_cardio["Session_Duration (hours)"], y = df_cardio["Calories_Burned"], s = 10 + df_water * 100, alpha = 0.7)
ax.set_xlabel("Session Duration (hours)")
ax.set_ylabel("Calories Burned")

# produce a legend with a cross-section of sizes from the scatter
ax.text(0.5, 1600, "Water (L)")

# find "normalized" values corresponding to 1.5, 2.5, and 3.5 liters
norm_vals = [(x - min_water) / (max_water - min_water) for x in [1.5, 2.5, 3.5]]
new_sizes = [10 + x * 100 for x in norm_vals]
ax.scatter([0.53] *3, [1530, 1465, 1400], s=new_sizes, alpha = 0.7, color="tab:blue")

# Add text labels 
ax.text(0.58, 1530, "1.5", va="center")
ax.text(0.58, 1465, "2.5", va="center")
ax.text(0.58, 1400, "3.5", va="center")



plt.show()
```

One more thing to add to this picture -- let's add colors to categorize "experience level" which has values 1, 2, or 3:

```python
fig, ax = plt.subplots()

def transform_colors(x):
    if x == 1:
        return "tab:blue"
    if x == 2:
        return "tab:orange"
    return "tab:green"

df_colors = df_cardio["Experience_Level"].apply(lambda x: transform_colors(x))

plot = ax.scatter(x = df_cardio["Session_Duration (hours)"], y = df_cardio["Calories_Burned"], s = 10 + df_water * 100, alpha = 0.7, c = df_colors)
ax.set_xlabel("Session Duration (hours)")
ax.set_ylabel("Calories Burned")

# produce a legend with a cross-section of sizes from the scatter
ax.text(0.5, 1600, "Water (L)")

# find "normalized" values corresponding to 1.5, 2.5, and 3.5 liters
norm_vals = [(x - min_water) / (max_water - min_water) for x in [1.5, 2.5, 3.5]]
new_sizes = [10 + x * 100 for x in norm_vals]
ax.scatter([0.53] *3, [1530, 1465, 1400], s=new_sizes, alpha = 0.7, color="black")

# Add text labels 
ax.text(0.58, 1530, "1.5", va="center")
ax.text(0.58, 1465, "2.5", va="center")
ax.text(0.58, 1400, "3.5", va="center")

# Add experience level labels
ax.text(0.6, 600, "Level 1", color ="tab:blue", fontweight="bold", ha = "center")
ax.text(1.3, 1300, "Level 2", color ="tab:orange", fontweight="bold", ha = "center")
ax.text(1.9, 900, "Level 3", color ="tab:green", fontweight="bold", ha = "center")



plt.show()
```
